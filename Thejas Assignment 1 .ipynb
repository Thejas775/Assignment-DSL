```python
def main():
    group_a=[]
    group_b=[]
    group_c=[]
    union=[]
    a=int(input('Enter the total number of students who play cricket '))
    for i in range(a):
        al=int(input('Enter the roll no of student {}  '.format(i+1)))
        group_a.append(al)
    print(group_a)
    b=int(input('Enter the total number of students who play badminton '))
    for i in range(b):
        bl=int(input('Enter the roll no of student {}  '.format(i+1)))
        group_b.append(bl)
    print(group_b)
    c=int(input('Enter the total number of students who play football '))
    for i in range(c):
        cl=int(input('Enter the roll no of student {}  '.format(i+1)))
        group_c.append(cl)
    print(group_c)
    d=int(input('Enter the total number of students who play nothing'))
    for i in range(d):
        d1=int(input('Enter the roll no of student {}  '.format(i+1)))
        union.append(d1)
    for each in group_a:
        union.append(each)
    for each in group_b:
        if each not in union:
            union.append(each)
    for each in group_c:
        if each not in union:
            union.append(each)
    print('Enter 1 to get list of students who play both cricket and badmintion')
    print('Enter 2 to get list of students who play either cricket or badmintion but not both')
    print('Enter 3 to get n of students who play neither cricket nor badmintion')
    print('Enter 4 to get number of students who play cricket and football but not badmintion ')
    number=int(input('Enter a number to choose a procedure'))
    



    def cricket_and_badmintion():
        print('List of students who play both cricket and badminton is ')
        l=[]
        for each in group_a:
            if each in group_b:
                l.append(each)
        print(l)
    def cricket_or_badmintion():
        print('List of students who play either cricket or badminton is ')
        l=[]
        for each in group_a:
            if each not in group_b:
                l.append(each)
        for each in group_b:
            if each not in group_a:
                l.append(each)
        print(l)

    def neither_cricket_nor_badmintion():
        print('Number of students who play neither cricket nor badminton is ')
        sum=0
        for each in union:
            if each not in group_b and each not in group_a:
                sum+=1
        print(sum)
    def both_cricket_and_football_but_not_badminton():
        print('Number of students who play both cricket and football but not badminton ')
        sum=0
        for each in group_a:
            if each in group_c and each not in group_b:
                sum+=1
        print(sum)
    if number==1:
        cricket_and_badmintion()
    if number==2:
        cricket_or_badmintion()
    if number==3:
        neither_cricket_nor_badmintion()
    if number==4:
        both_cricket_and_football_but_not_badminton()


main()
```
