```python
def main(choice):
    def make_matrix():
        m=int(input("Enter the number of columns: "))
        n=int(input("Enter the number of rows: "))
        mat=[]
        for i in range(n):
            mat.append([])
        for i in range(n):
            for j in range(m):
                mat[i].append(int(input("Enter the {}th column and {}th row element row wise: ".format(j+1,i+1))))
        return mat
    mat1=make_matrix()
    print("Your first matrix is: ")
    for each in mat1:
        print(each)
    mat2=make_matrix()
    print("Your Second matrix is: ")
    for each in mat2:
        print(each)
    def add(mat1,mat2):
        if len(mat1)==len(mat2) and len(mat1[0])==len(mat2[0]):
            add=[]
            for i in range(len(mat1)):
                add.append([])
            for i in range(len(mat1)):
                for j in range(len(mat2[0])):
                    add[i].append(mat1[i][j]+mat2[i][j])
        else:
            print("Addition is not possible due to different sizes of matirix")
            return
        print("Addition of matrices is: ")
        for each in add:
            print(each)
    def subtract(mat1,mat2):
        if len(mat1)==len(mat2) and len(mat1[0])==len(mat2[0]):
            add=[]
            for i in range(len(mat1)):
                add.append([])
            for i in range(len(mat1)):
                for j in range(len(mat2[0])):
                    add[i].append(mat1[i][j]-mat2[i][j])
        else:
            print("Subtraction is not possible due to different sizes of matirix")
            return
        print("Subtraction of matrices is: ")
        for each in add:
            print(each)
                
    def matmul(mat1,mat2):
        if len(mat1[0])==len(mat2):
            new=[]
            for i in range(len(mat1)):
                new.append([])
            for i in range(len(mat1)):
                for j in range(len(mat2[0])):
                    new[i].append(0)
            for i in range(len(mat1)):
                for j in range(len(mat2[0])):
                    for k in range(len(mat2)):
                        new[i][j] += mat1[i][k] * mat2[k][j]
            print(new)
        else:
            print("Dimension for multiplication does not match ")
    def transpose(mat1,mat2):
        new1=[]
        for i in range(len(mat1[0])):
            new1.append([])
        for i in range(len(mat1[0])):
            for j in range(len(mat1)):
                new1[i].append(0)
        for i in range(len(new1)):
            for j in range(len(new1[0])):
                new1[i][j]=mat1[j][i]
        for each in new1:
            print(each)
        new2=[]
        for i in range(len(mat2[0])):
            new2.append([])
        for i in range(len(mat2[0])):
            for j in range(len(mat2)):
                new2[i].append(0)
        for i in range(len(new2)):
            for j in range(len(new2[0])):
                new2[i][j]=mat2[j][i]
        for each in new2:
            print(each)
        
    if choice==1:
        add(mat1,mat2)
    if choice==2:
        subtract(mat1,mat2)
    if choice==3:
        matmul(mat1,mat2)
    if choice==4:
        transpose(mat1,mat2)
        
        
            
    
    
    
                    
        
        
    
    
n=int(input("Enter 1 to perform addition\nEnter 2 to perform subtraction\nEnter 3 to perform matrix multiplication\nEnter 4 to perform Transpose\n"))
main(n)        
```
